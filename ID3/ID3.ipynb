エントロピー

```python
def entropy(targetL):
  countc = {}
  total = 0
  for c in classnames:
    countc[c] = 0
  for f in targetL:
    countc[cattr[f]] += 1
    total += 1
  e = 0
  for c in classnames:
    if countc[c] > 0 and total > 0:
      e -= countc[c]/total * math.log2(countc[c]/total)
  return(e)
```

木構造

```python
# 木構造
class node():
  def __init__(self, inputL, myparent=None):
    self.L = copy.copy(inputL)
    self.child = []
    self.parent = myparent
    self.attr=None
    self.label=None
  def appendchild(self, inputL):
    newnode = node(inputL, self) # inputL：例題集合、self：親ノード
    self.child.append(newnode)
    return(newnode)
```

ジニ不純度

```python
def Gini(rootL,cattr):
  a=0
  b=0
  c=0
  d=0
  e=0
  for f in rootL:
    if cattr[f]==1:
      a+=1
    elif cattr[f]==2:
      b+=1
    elif cattr[f]==3:
      c+=1
    elif cattr[f]==4:
      d+=1
    elif cattr[f]==5:
      e+=1
  a/=len(rootL)
  b/=len(rootL)
  c/=len(rootL)
  d/=len(rootL)
  e/=len(rootL)
  # print(a,len(rootL))
  
  return 1-a*a-b*b-c*c-d*d-e*e
```
